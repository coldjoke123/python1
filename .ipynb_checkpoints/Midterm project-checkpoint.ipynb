# The ODE of the pendulum

The second-order ODE in the video of 3Blue1Brown is given as follow.
$$\theta'' = -\mu\theta' - \frac{g}{L}\sin(\theta)$$

where $\theta = \theta(t)$ is the angle of a pendulum to its lowest vertical state. The term $-\mu\theta'$ represents the air resistence which is proportional to the velocity. $g$ and $L$ are the acceleration of gravity and the length of the string of the pendulum, respectively.

# Procedure of simulation

The time interval has infinitely many points. However, we could barely simulate a finite number of points. Thus, for a time interval $[0,T]$, we choose an equidistant partition of the interval
$$0 = t_0 < t_1 <\cdots< t_{n-1} < t_n = T$$
where $t_i = i\Delta t$, for $i = 0,\cdots,n$. We now illustrate the function of the angle with the initial angle of $\theta(0) = \theta_0 = \frac{\pi}{3}$ and the initial angular velocity be $\theta'(0) = 0$.

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt

#physical constants
g = 9.8
L = 2
mu = 0.1

THETA_0 = np.pi/3 #60 degrees
THETA_DER_0 = 0 #No initial angular velocity

#Definition of ODE
def get_theta_sec_der(theta, theta_der):
    return -mu * theta_der - (g / L) * np.sin(theta)

#Solution to the differential equation
def theta(t):
    #Initialize changing values
    theta = THETA_0
    theta_der = THETA_DER_0
    delta_t = 0.01 # Some time step
    for time in np.arange(0, t, delta_t):
        theta_sec_der = get_theta_sec_der(theta, theta_der)
        theta += theta_der * delta_t
        theta_der += theta_sec_der * delta_t
    return theta
```

```python
x = np.linspace(0,20,500)
theta_sol = [(theta(t)/np.pi) for t in x]
plt.plot(x, theta_sol)
plt.xlabel('time')
plt.ylabel('angle');
```

The above result seems reasonable since the air resistence will slow down the pendulum and will eventually stop it.

# We may change the initial angular velocity to observe what will happen

Set the initial angular velocity to $\theta'(0) = -5$. That is, add some force to the pendulum toward the ground. We can see that the pendulum swing to a higher position.

```python
THETA_DER_0 = -4 #Set the initial angular velocity to -5
theta_sol = [(theta(t)/np.pi) for t in x]
plt.plot(x, theta_sol)
plt.xlabel('time')
plt.ylabel('angle');
```

# An example of the linear system of ODE

Given two species of animals, says rabbit and wolf. The relation between them is prey-and-predator. The ODE describes this relation is called the Lotka-Volterra equation, which is in the form of
$$r'=r(\alpha-\beta w)$$
$$w'=-w(\gamma-\delta r)$$

where $r = r(t)$ and $w = w(t)$ are the number of rabbits and wolves respectively. Also, the coefficients $\alpha,\beta,\gamma$ and $\delta$ are positive real parameters describing the interaction of the two species. For instance, let $\alpha = 4$, $\beta = 0.07$, $\gamma = 4$ and $\delta = 0.05$ as well as the initial number of rabbits and wolves be $r(0) = r_0 = 300$ and $w(0) = w_0 = 100$.
# Note that in this case. We didn't put the carrying capacity into consideration!

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt

alpha = 4
beta = 0.07
gamma = 4
delta = 0.05
r_0 = 300 #rabbits
w_0 = 100 #wolves

#Definition of ODE
def get_r_der(r, w):
    return r*(alpha - beta * w)
def get_w_der(r, w):
    return -w*(gamma - delta * r)

#Solution to the differential equation
def LVsol(t):
    #Initialize changing values
    r = r_0
    w = w_0
    delta_t = 0.01 # Some time step
    for time in np.arange(0, t, delta_t):
        r_der = get_r_der(r, w)
        w_der = get_w_der(r, w)
        r += r_der * delta_t
        w += w_der * delta_t
    return (r, w)
```

```python
x = np.linspace(0,15,200)
r_sol = zip(*[LVsol(t) for t in x])[0]
w_sol = zip(*[LVsol(t) for t in x])[1]
plt.plot(x, r_sol)
plt.plot(x, w_sol)
plt.legend(['rabbit','wolf'], loc = 'best')
plt.xlabel('time')
plt.ylabel('number')
plt.title('Lotka-Volterra equation');
```
